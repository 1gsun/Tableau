```python
import pandas as pd 
```

```python
tb = pd.read_csv("657516099_T_ONTIME.csv")
loc = pd.read_csv("airports.dat.txt",header = -1)
loc = loc[[4,6,7]]
tb1 = pd.merge(tb,loc,how = "left", left_on = tb["ORIGIN"],right_on = loc[4])
tb1.rename(columns={4: 'ORG_ID', 6: 'ORG_LAT',7:'ORG_LOG'}, inplace=True)
tb2 = pd.merge(tb1,loc,how = "left", left_on = tb["DEST"],right_on = loc[4])
tb2.rename(columns={4: 'DEST_ID', 6: 'DEST_LAT',7:'DEST_LOG'}, inplace=True)
```

```python
def crs_dep(tb2):
    return str(tb2["CRS_DEP_TIME"])[:-2]+":"+str(tb2["CRS_DEP_TIME"])[-2:]

tb2["CRS_DEP_TIME"] = tb2.apply(crs_dep,axis = 1)
   
```

```python
def crs_arr(tb2):
    return str(tb2["CRS_ARR_TIME"])[:-2]+":"+str(tb2["CRS_ARR_TIME"])[-2:]

tb2["CRS_ARR_TIME"] = tb2.apply(crs_arr,axis = 1)
```

```python
def path(tb2):
    return (tb2["ORG_ID"]+"-"+tb2["DEST_ID"])

tb2["PATH"] = tb2.apply(path,axis = 1)
```

```python
import copy
```

```python
tb3 = copy.deepcopy(tb2)
tb3["org_dest"] = "Origin"
tb3["station"] = tb3["ORG_ID"]
tb3["path_id "] = tb3["PATH"]
tb3["latitude"] = tb3["ORG_LAT"]
tb3["longitude"] = tb3["ORG_LOG"]
```

```python
tb4 = copy.deepcopy(tb2)
tb4["org_dest"] = "Destination"
tb4["station"] = tb4["DEST_ID"]
tb4["path_id "] = tb4["PATH"]
tb4["latitude"] = tb4["DEST_LAT"]
tb4["longitude"] = tb4["DEST_LOG"]
```

```python
tb_final =  pd.concat([tb3, tb4])
```

```python
tb_final.to_csv("Processed_data.csv")
```
